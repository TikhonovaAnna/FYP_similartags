```python
f = open('ManualCategory.txt',encoding = 'utf8')
lines = f.readlines()
print(len(lines))
```

```python
import re
cate = re.sub('[\',]',' ',lines[0])
```

```python
cate_indi = cate.split()
```

```python
cate_indi
```

```python
cate_no_line = []
for word in cate_indi:
    if '|' not in word:
        cate_no_line.append(word)
print(cate_no_line)
```

```python
cate_no_special = []
for word in cate_no_line:
    if re.match("^[a-zA-Z0-9_]*$", word) and len(word)>=3:
        cate_no_special.append(word)
```

```python
print(cate_no_special)
```

```python
cate_with_s = []
cate_no_s = []
for word in cate_no_special:
    if word.endswith('s'):
        cate_with_s.append(word)
    else:
        cate_no_s.append(word)
print(cate_with_s)
```

```python
set(cate_no_s)
fw = open('manualCategory.txt','w',encoding = 'utf8')
for word in cate_no_s:
    fw.write(word+ '\n')
fw.close()
```

```python
print(cate_no_s)
```

```python
import csv
csvfile = 'manualCategory.csv'
with open(csvfile,'w') as output:
    writer = csv.writer(output,lineterminator = '\n')
    for word in cate_no_s:
        writer.writerow([word,''])
```
